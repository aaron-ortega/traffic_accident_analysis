```python
# QUERY
# Create tables
for c in commands:
    cur.execute(c)

conn.commit()

# Create copy statements
tables = 'charges,crash,damages,endorsements,lookup,person,restrictions,unit'.split(',')
copy_tables_csv = []
for table, path in zip(tables, paths):
    copy_tables_csv.append(
        f'''
        COPY {table}
        FROM '{path}'
        DELIMITER ','
        CSV
        HEADER;
        ''')
    
copy_tables_csv = tuple(copy_tables_csv)

# QUERY
# Run copy query
for copy_query in copy_tables_csv:
    cur.execute(copy_query)
    
conn.commit()

# Finish Up
cur.close()
conn.close()
```
